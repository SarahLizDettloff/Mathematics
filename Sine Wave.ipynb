```python
import numpy
import matplotlib.pyplot as mathplot
```

```python
# When phase is not zero, then the entire sine wave plot will be shifted by the amount in time.
```

```python
phase = 0 
```

```python
# Ordinary Frequency is the number of cycles which will occur each second of time
```

```python
ordinary_frequency = 20
```

```python
# Also known as angular speed, radial frequency, circular frequency, orbital frequency, radian frequency, and pulsatance
# Angular Frequency is the rate of change of the function argument in units of radians per a second
```

```python
angular_frequency = 0.3
```

```python
time = numpy.arange(phase, ordinary_frequency, angular_frequency); 
```

```python
amplitude = numpy.sin(time)
```

```python
mathplot.plot(time, amplitude) # Plots time and aplitude
mathplot.title('Sine Wave') # Names the title above the graph
mathplot.xlabel('Time') # Names the x-axis
mathplot.ylabel('Amplitude') # Names the y-axis
mathplot.axhline(y=0, color='g') # Creates a line across the x-axis, set in the middle at y=0, and colors it green.
mathplot.grid(True, which='both') # Creates a gray grid behind the graph
mathplot.show() # Displays the graph
```
